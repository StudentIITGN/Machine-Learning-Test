Q.2)solve using list of list and numpy,compare time for both.

```python
#lets consider 2 array
#2*3  and  3*3= 2*3

List1=[[2,3],[4,5]]
List2=[[2,6,7],[8,5,9]]

newarr1=[[0,0,0],[0,0,0]]
for i in range(0,len(newarr1)):
    for j in range(0,len(newarr1[0])):
        for k in range(0,len(List2)):
            newarr1[i][j]+=List1[i][k]*List2[k][j]
for row in newarr1:
    print(row)

%timeit for row in newarr1:row
```

```python
import numpy as np

arr1=np.array([[2,3],[4,5]])
arr2=np.array([[2,6,7],[8,5,9]])
newarr1=np.matmul(arr1,arr2)
#newarr=np.dot(arr1,arr2)
print(newarr1)
%timeit newarr1
```

By seeing both time solution numpy array faster than list because Numpy arrays provides much more efficient storage and data operation than python lists.

Q.5)Use Jax to confirm solution

```python
import jax
import numpy as np

def func(x,y):
    return x**2 *y+y**3 *jax.numpy.sin(x)

x_value=3.0     #taking random value to compare solution
y_value=3.0

#solution using jax

grad_func=jax.grad(func,argnums=(0,1))
gradient=grad_func(x_value,y_value)
print(gradient)

#analytical solution

grad=np.array([2*x_value*y_value+y_value**3 *np.cos(x_value),x_value**2+3*y_value**2 *np.sin(x_value)])
print(grad)
```

Q.8) Store same information using python class.

```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks  # Assuming marks is a dictionary of subject-marks pairs

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

database = []

# Example 
year1 = Year(2022)
branch1= Branch('Earth Science')
branch1.add_student(Student('Dakshata', 1, {'Math':100,'English':90}))
branch1.add_student(Student('Ashwini', 2, {'Math':68,'English':72}))

year1.add_branch(branch1)

database.append(year1)

print(database[0].branches[0].students[0].name) 
print(database[0].branches[0].students[0].marks['English'])  
```

Q.13) function to calculate argmin of numpy array

```python
def calculate_minIndex(arr):
        if len(arr)==0:
            raise ValueError("array is empty")
            
        min_index=0
        min_value=arr[0]
        for i in range(1,len(arr)):
            if arr[i]<min_value:
                min_value=arr[i]
                min_index+=i
        return min_index
arr=[3,5,4,8,9,6,7,5,1]
minIndex1=calculate_minIndex(arr)
print(minIndex1)


minIndex2=np.argmin(arr)     #to verify solution
print(minIndex2)
```
